```python
import joblib
import pandas as pd
from pathlib import Path

# Load the saved preprocessor
model_path = Path("../models/preprocessor.pkl")
preprocessor = joblib.load(model_path)

print(f"Preprocessor loaded: {type(preprocessor)}")

# Inspect the Numerical Scaler (StandardScaler)
# This shows the 'mean' calculated during training
num_scaler = preprocessor.named_transformers_['num_pipeline']
print(f"Mean values learned: {num_scaler.mean_}")

# Inspect the Categorical Encoder (OneHotEncoder)
cat_encoder = preprocessor.named_transformers_['cat_pipeline']
print(f"Categories learned for 'sex': {cat_encoder.categories_[0]}")
```
